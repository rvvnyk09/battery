```python
import pandas as pd
import os
```

```python
folder_path = 'H:\\RV\\Work\\py\\Battery\\Dataset'

files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

dataframes = []
for f in files:
    df = pd.read_csv(os.path.join(folder_path, f), delimiter=';',encoding="latin")
    df['Distance(km)'] = (df['Velocity [km/h]'] / 3600) * df['Time [s]']
    df.insert(0, 'file_name', f)
    dataframes.append(df)

final_dataframe = pd.concat(dataframes, ignore_index=True, sort=False)

final_dataframe.to_csv('H:\\RV\\Work\\py\\Battery\\file.csv', index=False)

final_dataframe = final_dataframe.drop('file_name',axis=1)
final_dataframe = final_dataframe.drop('displayed SoC [%]',axis=1)
final_dataframe = final_dataframe.drop('Unnamed: 23',axis=1)
```

```python
# pearson's correlation feature selection for numeric input and numeric output
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
x = final_dataframe.drop('Distance(km)', axis=1)
y = final_dataframe['Distance(km)']
```

```python
# define feature selection
fs = SelectKBest(score_func=f_regression, k=10)
# apply feature selection
x_selected = fs.fit_transform(x, y)
```

```python
x_selected.shape
```

```python
# Get scores
scores = fs.scores_

# Assuming x is a dataframe and columns are feature names
feature_names = x.columns

# Combine feature names and scores into a DataFrame
feature_scores_df = pd.DataFrame({'Feature Name': feature_names, 'Score': scores})

# Sort the DataFrame by score in descending order
sorted_feature_scores = feature_scores_df.sort_values(by='Score', ascending=False)

# Display the top 10 features
top_10_features = sorted_feature_scores.head(10)
print(top_10_features)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
```

```python
# Assuming you have feature matrix X and target vector y
x_train, x_test, y_train, y_test = train_test_split(x_selected, y, test_size=0.2, random_state=42)

# Initialize the model
model = LinearRegression()

# Fit the model
modelfit = model.fit(x_train, y_train)
```

```python
# Predict on the test data
y_pred = model.predict(x_test)

# Calculate RMSE and R squared score
rme = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f'Mean_absolute_error: {rme}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R2 Score: {r2}')
```
